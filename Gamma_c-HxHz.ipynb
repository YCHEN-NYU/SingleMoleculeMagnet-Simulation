# $\Gamma(H_z, H_x)$ Calculation by Ferran's Method

```python
# calculation of Gamma from PHYSICAL REVIEW B 79, 092403 !2009" by Ferran
# take levels of m' = 8, 9, 10

import numpy as np
from scipy.misc import factorial
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import rc

def E(m, Hz):
	return -D*m**2 - Hz*m*muB*g + const
def delta(D, S, m, mp, Hx):
	return 2*D*(factorial(S + mp)*factorial(S-m)/factorial(S - mp)/factorial(S + m))**(0.5)* \
    (Hx*muB*g/(2*D))**(mp - m)/(factorial(abs(mp - m - 1)))**2
def tau_m(delta_mmp, gamma_mp, hbar, omega_mmp):
    return delta_mmp**2*(gamma_mp/2)/(2*hbar**2)/(omega_mmp**2+(gamma_mp/2)**2)

LabelX = '$H_z (T)$'
LabelY = '$H_x (T) $'
TitleName = '$log_{10}\Gamma (H_z, H_x)$'

fontNum = 28
fontAxis = 28
fontLegend = 28
fontTitle = 32
fontTick = 24
MKsize = 10

g = 1.99
muB = 9.274009994e-24 # =J/T
D = 0.708*1.380650e-23
S = 10
m = -10
hbar = 1.054571800e-34 # J*s
const = 0.0

Hzmin = 0
Hzmax = 1.25
Hxmin = 1.0
Hxmax = 5.0
clim_min = -10
clim_max = 10
NumofPoints = 1000
ct_intervals = 10

Hz = np.linspace(Hzmin, Hzmax, NumofPoints)
Hx = np.linspace(Hxmin, Hxmax, NumofPoints)
[Hzmesh, Hxmesh] = np.meshgrid(Hz, Hx)
Tau = np.zeros((len(Hz), len(Hx)))
S = 10
m = -10
```

```python
# figure properties
fig = plt.figure(figsize=(16, 10))
fig.patch.set_facecolor('white')
ax = plt.subplot(111)
# if mp==7:
#         gamma_mp=1.616867386364779e12/1000
#     elif mp == 8:
#         gamma_mp=1.616867386364779e12/20
#     else:
#         gamma_mp=1.616867386364779e12/20  
# take levels of m' = 8, 9, 10
for mp in range(8, 11):
    gamma_mp=1.616867386364779e12/20 
    omega_mmp=(E(m, Hzmesh)-E(mp, Hzmesh))/hbar
    delta_mmp = delta(D, S, m, mp, Hxmesh)
    Tau += tau_m(delta_mmp, gamma_mp, hbar, omega_mmp)
    
ct = plt.contour(Hz, Hx, np.log10(Tau), ct_intervals, color='k', linestype='--', linewidth=2)    
im = plt.pcolormesh(Hz, Hx, np.log10(Tau), cmap = plt.cm.get_cmap("rainbow"))
plt.colorbar(im)
plt.clim(clim_min, clim_max)
ax.set_xlim(Hzmin, Hzmax)
ax.set_ylim(Hxmin, Hxmax)
plt.rc('font', size=fontNum)
ax.set_frame_on
plt.rc('font', size=32)
plt.xlabel(LabelX, fontsize=fontAxis, fontweight='bold')
plt.ylabel(LabelY, fontsize=fontAxis, fontweight='bold')
plt.title(TitleName, fontsize=fontTitle, fontweight='bold')
ax.xaxis.set_tick_params(labelsize=fontTick)
ax.yaxis.set_tick_params(labelsize=fontTick)
plt.savefig('Gamma_HzHx_mp8to10.png')
plt.show()
```
