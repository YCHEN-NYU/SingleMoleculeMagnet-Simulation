```python
# theory - simulation of SMMs deflagration
# PHYSICAL REVIEW B 76, 054410  2007 Dmitry A. Garanin

import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
u = np.linspace(-50, 10, 1000)
n = (1 - np.tanh(u/2.0))/2

# plt.figure(figsize=(10,10))
plt.plot(u, n, color='g', linewidth = 5)
plt.xlabel('coordinate u', fontsize=20)
plt.ylabel('reduced population n', fontsize=20)
plt.title('n-u', fontsize=20)
plt.grid('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
def mandelbrot( h,w, maxit=50):
    """Returns an image of the Mandelbrot fractal of size (h,w)."""
    y,x = np.ogrid[ -1.4:1.4:h*1j, -2:0.8:w*1j ]
    c = x+y*1j
    z=c
    divtime = maxit + np.zeros(z.shape, dtype=int)
    for i in range(maxit):
        z=z**4+c
        diverge = z*np.conj(z) > 2**2 # who is diverging
        div_now = diverge & (divtime==maxit) # who is diverging now
        divtime[div_now] = i
        z[diverge] = 2
    return divtime
plt.figure(figsize = (10, 10))
plt.imshow(mandelbrot(400,400))
plt.colorbar()
plt.show()
```

```python
# SH approach 
# E = -DS^2 + mu*g*Hz*S
import numpy as np
import matplotlib.pyplot as plt
D = -0.5
mu = 1
g = 2
Hz = np.linspace(0, 5, 100)
S = range(-10, 10)
fig = plt.figure(figsize = (8, 6))
for sz in S: 
    Energy = D*sz**2 + mu*g*Hz*sz
    plt.plot(Hz, Energy)
plt.xlim([0, 5])
plt.ylim([-150, 50])
plt.xlabel('$H_z (T)$')
plt.ylabel('$E(K)$')
plt.show()
```
